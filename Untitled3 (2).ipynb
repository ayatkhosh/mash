```python
# ═══════════════════════════════════════════════
# CELL 4 — تحميل النماذج (Fallback آمن بدون تعديل الملفات)
# ═══════════════════════════════════════════════

import sys, os
import torch
import torch.nn.functional as F

os.environ["TOKENIZERS_PARALLELISM"] = "false"
if "/content/GroundingDINO" not in sys.path:
    sys.path.insert(0, "/content/GroundingDINO")

# ───────────────────────────────────────────────
# 1) Patch ms_deform_attn FORWARD in-memory only
# ───────────────────────────────────────────────
def ms_deform_attn_forward_pytorch(value, value_spatial_shapes, sampling_locations, attention_weights):
    """
    نفس الدالة الموجودة أصلاً داخل الملف الرسمي:
    multi_scale_deformable_attn_pytorch لكن بشكل مستقل هنا.
    """
    bs, _, num_heads, embed_dims = value.shape
    _, num_queries, num_heads, num_levels, num_points, _ = sampling_locations.shape

    value_list = value.split([H_ * W_ for H_, W_ in value_spatial_shapes], dim=1)
    sampling_grids = 2 * sampling_locations - 1
    sampling_value_list = []

    for level, (H_, W_) in enumerate(value_spatial_shapes):
        value_l_ = value_list[level].flatten(2).transpose(1, 2).reshape(bs * num_heads, embed_dims, H_, W_)
        sampling_grid_l_ = sampling_grids[:, :, :, level].transpose(1, 2).flatten(0, 1)
        sampling_value_l_ = F.grid_sample(
            value_l_, sampling_grid_l_,
            mode="bilinear", padding_mode="zeros", align_corners=False
        )
        sampling_value_list.append(sampling_value_l_)

    attention_weights = attention_weights.transpose(1, 2).reshape(
        bs * num_heads, 1, num_queries, num_levels * num_points
    )
    output = (
        (torch.stack(sampling_value_list, dim=-2).flatten(-2) * attention_weights)
        .sum(-1)
        .view(bs, num_heads * embed_dims, num_queries)
    )
    return output.transpose(1, 2).contiguous()

def patch_ms_deform_attn():
    import groundingdino.models.GroundingDINO.ms_deform_attn as ms

    # نأخذ الـ Function class من الملف
    MSFunc = ms.MultiScaleDeformableAttnFunction

    # نبدل forward فقط (inference). backward غير مهم لأننا نعمل no_grad()
    @staticmethod
    def forward(ctx, value, value_spatial_shapes, value_level_start_index,
                sampling_locations, attention_weights, im2col_step):
        # ignore level_start_index + im2col_step هنا، لأنها تخص C++ impl
        return ms_deform_attn_forward_pytorch(
            value, value_spatial_shapes, sampling_locations, attention_weights
        )

    MSFunc.forward = forward
    print("✅ Patched MultiScaleDeformableAttnFunction.forward to pure PyTorch (no _C).")

print("🔧 Applying safe CPU/GPU fallback patch...")
patch_ms_deform_attn()

# ───────────────────────────────────────────────
# 2) تحميل GroundingDINO + SAM
# ───────────────────────────────────────────────
print("\n📥 [1/2] تحميل Grounding DINO...")

DINO_CONFIG_URL  = "https://raw.githubusercontent.com/IDEA-Research/GroundingDINO/main/groundingdino/config/GroundingDINO_SwinB_cfg.py"
DINO_WEIGHTS_URL = "https://github.com/IDEA-Research/GroundingDINO/releases/download/v0.1.0-alpha2/groundingdino_swinb_cogcoor.pth"

DINO_CONFIG  = DIRS["models"] / "groundingdino_config.py"
DINO_WEIGHTS = DIRS["models"] / "groundingdino_swinb.pth"

if not DINO_CONFIG.exists():
    os.system(f'wget -q "{DINO_CONFIG_URL}" -O "{DINO_CONFIG}"')
print("  ✅ ملف الإعداد جاهز")

if not DINO_WEIGHTS.exists():
    print("  📥 جاري تحميل أوزان DINO (~938 MB)...")
    os.system(f'wget -q "{DINO_WEIGHTS_URL}" -O "{DINO_WEIGHTS}"')
    print("  ✅ أوزان DINO جاهزة")
else:
    print("  ✅ أوزان DINO موجودة مسبقاً")

from groundingdino.util.inference import load_model as load_dino
from groundingdino.util.inference import predict as dino_predict
from groundingdino.util.inference import load_image as dino_load_image

dino_model = load_dino(str(DINO_CONFIG), str(DINO_WEIGHTS))
dino_model = dino_model.to(device)
dino_model.eval()
print(f"  ✅ Grounding DINO جاهز على {device.upper()}")

print("\n📥 [2/2] تحميل SAM...")

SAM_WEIGHTS_URL = "https://dl.fbaipublicfiles.com/segment_anything/sam_vit_h_4b8939.pth"
SAM_WEIGHTS     = DIRS["models"] / "sam_vit_h.pth"

if not SAM_WEIGHTS.exists():
    print("  📥 جاري تحميل أوزان SAM (~2.4 GB)...")
    os.system(f'wget -q "{SAM_WEIGHTS_URL}" -O "{SAM_WEIGHTS}"')
    print("  ✅ أوزان SAM جاهزة")
else:
    print("  ✅ أوزان SAM موجودة مسبقاً")

from segment_anything import sam_model_registry, SamPredictor

sam_model     = sam_model_registry["vit_h"](checkpoint=str(SAM_WEIGHTS))
sam_model     = sam_model.to(device)
sam_predictor = SamPredictor(sam_model)
print(f"  ✅ SAM ViT-H جاهز على {device.upper()}")

print("\n" + "="*50)
print("🚀 جميع النماذج جاهزة للعمل!")
print("="*50)
```

```python
# ═══════════════════════════════════════════════
# CELL 5 — تحميل Dataset
# ═══════════════════════════════════════════════

# ─── محاولة تحميل INRIA Dataset ───
# هذا الداتاسيت ضخم، سنحمّل جزءاً منه فقط (50-100 صورة)

print('📂 جاري إعداد Dataset...')
print('='*50)

def download_sample_dataset(target_dir, n_images=60):
    """
    تحميل عينة من صور جوية من مصادر مفتوحة.
    نستخدم مجموعة SpaceNet كبديل خفيف للـ INRIA.
    """
    target_dir = Path(target_dir)
    images_dir = target_dir / 'images'
    masks_dir  = target_dir / 'masks'
    images_dir.mkdir(exist_ok=True)
    masks_dir.mkdir(exist_ok=True)

    # قائمة صور جوية مفتوحة المصدر (عينة من DOTA / Potsdam)
    # في مشروع حقيقي: استخدم INRIA من https://project.inria.fr/aerialimagelabeling/
    sample_urls = [
        'https://upload.wikimedia.org/wikipedia/commons/thumb/1/1b/Colosseum_in_Rome%2C_Italy_-_April_2007.jpg/1280px-Colosseum_in_Rome%2C_Italy_-_April_2007.jpg',
    ]

    # بدلاً من ذلك، نولّد صور اصطناعية واقعية للاختبار
    print(f'  🖼️  توليد {n_images} صورة تجريبية...')

    generated = []
    for i in tqdm(range(n_images), desc='توليد الصور'):
        # نولّد صورة جوية اصطناعية بألوان وأشكال تحاكي المدينة
        h, w = 512, 512
        img  = np.ones((h, w, 3), dtype=np.uint8) * 180  # خلفية رمادية (إسفلت)
        mask = np.zeros((h, w), dtype=np.uint8)           # قناع المباني

        np.random.seed(SEED + i)

        # نضيف أرضية عشوائية
        noise = np.random.randint(0, 30, (h, w, 3), dtype=np.uint8)
        img   = np.clip(img.astype(int) + noise - 15, 0, 255).astype(np.uint8)

        # نرسم مناطق خضراء
        n_green = np.random.randint(2, 5)
        for _ in range(n_green):
            gx, gy = np.random.randint(0, w-80), np.random.randint(0, h-80)
            gw, gh = np.random.randint(40, 120), np.random.randint(40, 120)
            green_color = np.array([np.random.randint(30, 80),
                                    np.random.randint(100, 170),
                                    np.random.randint(20, 60)])
            img[gy:gy+gh, gx:gx+gw] = green_color

        # نرسم مبانٍ
        n_buildings = np.random.randint(5, 15)
        for _ in range(n_buildings):
            bx = np.random.randint(0, w - 60)
            by = np.random.randint(0, h - 60)
            bw = np.random.randint(30, 80)
            bh = np.random.randint(30, 80)
            # لون المبنى (رمادي دافئ أو بني)
            c  = np.random.randint(120, 220)
            building_color = np.array([c, c-20, c-40]).clip(0, 255)
            img[by:by+bh, bx:bx+bw]  = building_color
            mask[by:by+bh, bx:bx+bw] = 255  # نسجّل المبنى في القناع

            # نضيف ظل بسيط (مثل المباني الحقيقية)
            shadow_w = max(3, bw//4)
            sx2 = min(bx + bw + shadow_w, w)
            img[by:by+bh, bx+bw:sx2] = np.array([80, 80, 70])

        # نحفظ الصورة والقناع
        img_path  = images_dir / f'aerial_{i:04d}.jpg'
        mask_path = masks_dir  / f'aerial_{i:04d}.png'
        Image.fromarray(img).save(img_path)
        Image.fromarray(mask).save(mask_path)
        generated.append({'image': str(img_path), 'mask': str(mask_path)})

    return generated


# ─── تحميل أو توليد البيانات ───
all_data = download_sample_dataset(DIRS['data'], n_images=100)

# ─── تقسيم البيانات ───
# 70% train, 15% val, 15% test
train_data, temp_data = train_test_split(all_data, test_size=0.30, random_state=SEED)
val_data,   test_data = train_test_split(temp_data, test_size=0.50, random_state=SEED)

print(f'\n📊 إحصائيات Dataset:')
print(f'   إجمالي الصور : {len(all_data)}')
print(f'   Train         : {len(train_data)} صورة (70%)')
print(f'   Validation    : {len(val_data)} صورة (15%)')
print(f'   Test          : {len(test_data)} صورة (15%)')
print(f'\n✅ Dataset جاهز!')

# نعرض عينة
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle('عينة من Dataset (صور + أقنعة)', fontsize=14, fontweight='bold')
for idx, ax_row in enumerate(axes.T):
    sample = train_data[idx]
    img  = np.array(Image.open(sample['image']))
    mask = np.array(Image.open(sample['mask']))
    ax_row[0].imshow(img);  ax_row[0].set_title(f'صورة {idx}', fontsize=10)
    ax_row[1].imshow(mask, cmap='gray'); ax_row[1].set_title(f'قناع {idx}', fontsize=10)
    for ax in ax_row: ax.axis('off')
plt.tight_layout()
plt.savefig(DIRS['vis'] / 'dataset_sample.png', dpi=100, bbox_inches='tight')
plt.show()
```

```python
# ═══════════════════════════════════════════════
# CELL 7 — الكشف بـ Grounding DINO
# ═══════════════════════════════════════════════

from groundingdino.util.inference import predict as dino_predict

def run_dino_detection(image_path: str, prompts: dict, thresholds: dict) -> dict:
    """
    يُشغّل Grounding DINO على صورة واحدة لكل الفئات.

    المدخلات:
        image_path : مسار الصورة
        prompts    : قاموس {فئة: نص_البحث}
        thresholds : قاموس {فئة: {box, text}}

    المخرجات:
        قاموس {فئة: {'boxes': ..., 'logits': ..., 'phrases': ...}}
    """
    # تحميل الصورة بتنسيق DINO
    image_source, image_tensor = dino_load_image(image_path)
    h, w = image_source.shape[:2]

    results = {}

    for category, prompt_text in prompts.items():
        thresh = thresholds[category]

        with torch.no_grad():
            boxes, logits, phrases = dino_predict(
                model      = dino_model,
                image      = image_tensor,
                caption    = prompt_text,
                box_threshold  = thresh['box'],
                text_threshold = thresh['text'],
                device     = device
            )

        # تحويل الصناديق من نسبية [0,1] إلى بكسل
        # DINO يُرجع [cx, cy, w, h] نسبية → نحوّل لـ [x1, y1, x2, y2] بكسل
        if len(boxes) > 0:
            boxes_np = boxes.cpu().numpy()
            cx, cy, bw, bh = boxes_np[:,0], boxes_np[:,1], boxes_np[:,2], boxes_np[:,3]
            x1 = ((cx - bw/2) * w).clip(0, w).astype(int)
            y1 = ((cy - bh/2) * h).clip(0, h).astype(int)
            x2 = ((cx + bw/2) * w).clip(0, w).astype(int)
            y2 = ((cy + bh/2) * h).clip(0, h).astype(int)
            pixel_boxes = np.stack([x1, y1, x2, y2], axis=1)
        else:
            pixel_boxes = np.empty((0, 4), dtype=int)

        results[category] = {
            'boxes'   : pixel_boxes,
            'scores'  : logits.cpu().numpy() if len(logits) > 0 else np.array([]),
            'phrases' : phrases,
            'image_wh': (w, h)
        }

    return results, image_source


def apply_nms(boxes: np.ndarray, scores: np.ndarray, iou_threshold: float = 0.5) -> np.ndarray:
    """
    Non-Maximum Suppression — يُزيل الصناديق المتداخلة المكررة.
    يحتفظ فقط بالصندوق ذي الثقة الأعلى في كل تجمّع.
    """
    if len(boxes) == 0:
        return np.array([], dtype=int)

    x1, y1, x2, y2 = boxes[:,0], boxes[:,1], boxes[:,2], boxes[:,3]
    areas = (x2 - x1 + 1) * (y2 - y1 + 1)
    order = scores.argsort()[::-1]

    keep = []
    while order.size > 0:
        i = order[0]
        keep.append(i)

        xx1 = np.maximum(x1[i], x1[order[1:]])
        yy1 = np.maximum(y1[i], y1[order[1:]])
        xx2 = np.minimum(x2[i], x2[order[1:]])
        yy2 = np.minimum(y2[i], y2[order[1:]])

        inter = np.maximum(0, xx2 - xx1 + 1) * np.maximum(0, yy2 - yy1 + 1)
        iou   = inter / (areas[i] + areas[order[1:]] - inter)

        order = order[np.where(iou <= iou_threshold)[0] + 1]

    return np.array(keep)


# ─── اختبار على صورة واحدة من Test Set ───
print('🔍 اختبار الكشف على صورة واحدة...')
test_image_path = test_data[0]['image']

detection_results, test_image = run_dino_detection(
    image_path = test_image_path,
    prompts    = BEST_PROMPTS,
    thresholds = THRESHOLDS
)

print('\n📊 نتائج الكشف:')
total = 0
for cat, res in detection_results.items():
    n = len(res['boxes'])
    total += n
    emoji = {'buildings': '🏢', 'vegetation': '🌳', 'open_land': '🏜️'}[cat]
    print(f'  {emoji} {cat}: {n} كشف')
print(f'  📌 المجموع: {total} كشف')

print('\n✅ الكشف يعمل بنجاح!')
print('   (ملاحظة: الصور التجريبية البسيطة قد تعطي كشوفات قليلة — هذا طبيعي)')
print('   (مع صور جوية حقيقية ستكون النتائج أغنى بكثير)')
```

```python
# ═══════════════════════════════════════════════
# CELL 8 — التقسيم بـ SAM
# ═══════════════════════════════════════════════

def run_sam_segmentation(image_np: np.ndarray, boxes: np.ndarray) -> List[np.ndarray]:
    """
    يُشغّل SAM على مجموعة صناديق لصورة واحدة.

    المدخلات:
        image_np : صورة NumPy بصيغة RGB (H, W, 3)
        boxes    : مصفوفة صناديق (N, 4) بصيغة [x1,y1,x2,y2]

    المخرجات:
        قائمة بالأقنعة — كل قناع مصفوفة Boolean (H, W)
    """
    if len(boxes) == 0:
        return []

    # 1. نُسجّل الصورة في SAM (يُحوّلها لـ embedding)
    sam_predictor.set_image(image_np)

    masks_list = []

    for box in boxes:
        x1, y1, x2, y2 = box

        # تجاهل الصناديق الصغيرة جداً (ضوضاء)
        if (x2 - x1) < 5 or (y2 - y1) < 5:
            continue

        # 2. نُمرّر الصندوق لـ SAM ونحصل على القناع
        input_box = np.array([x1, y1, x2, y2])

        with torch.no_grad():
            masks, scores, _ = sam_predictor.predict(
                point_coords = None,
                point_labels = None,
                box          = input_box[None, :],  # SAM يتوقع batch dimension
                multimask_output = True              # يُرجع 3 أقنعة، نختار الأفضل
            )

        # 3. نختار القناع ذا الثقة الأعلى
        best_mask_idx = scores.argmax()
        best_mask     = masks[best_mask_idx]  # Boolean array (H, W)
        masks_list.append(best_mask)

    return masks_list


def create_land_use_map(image_shape: Tuple, detection_results: dict) -> np.ndarray:
    """
    FR-4: ينشئ خريطة استخدامات الأرض.

    الفئات:
        0 = open_land  (أرض مفتوحة)
        1 = vegetation (أخضر)
        2 = built_up   (مبني)

    المخرجات:
        خريطة (H, W) بأرقام الفئات
    """
    h, w = image_shape[:2]
    land_map = np.zeros((h, w), dtype=np.uint8)  # كل شيء أرض مفتوحة بداية

    # طبقة 1: المناطق الخضراء (تحت المباني)
    for mask in detection_results.get('vegetation_masks', []):
        land_map[mask] = 1

    # طبقة 2: المباني (فوق الخضراء — أولوية أعلى)
    for mask in detection_results.get('buildings_masks', []):
        land_map[mask] = 2

    return land_map


# ─── تشغيل SAM على نتائج DINO ───
print('✂️  تشغيل SAM على نتائج الكشف...')

# تحضير الصورة
test_image_rgb = test_image  # الصورة من Cell 7

all_seg_results = {}

for category in ['buildings', 'vegetation', 'open_land']:
    boxes  = detection_results[category]['boxes']
    scores = detection_results[category]['scores']

    # تطبيق NMS لإزالة الصناديق المتداخلة
    if len(boxes) > 0 and len(scores) > 0:
        keep_idx = apply_nms(boxes, scores, iou_threshold=0.45)
        boxes    = boxes[keep_idx]

    emoji = {'buildings': '🏢', 'vegetation': '🌳', 'open_land': '🏜️'}[category]
    print(f'  {emoji} {category}: {len(boxes)} صندوق بعد NMS → جاري التقسيم...')

    # تشغيل SAM
    masks = run_sam_segmentation(test_image_rgb, boxes)

    all_seg_results[f'{category}_boxes'] = boxes
    all_seg_results[f'{category}_masks'] = masks

    print(f'       → {len(masks)} قناع مُولَّد')


# ─── إنشاء خريطة استخدامات الأرض (FR-4) ───
land_use_map = create_land_use_map(test_image_rgb.shape, all_seg_results)

# ─── حساب النسب المئوية ───
total_pixels    = land_use_map.size
n_open     = np.sum(land_use_map == 0)
n_green    = np.sum(land_use_map == 1)
n_built    = np.sum(land_use_map == 2)

print('\n📊 تقسيم استخدامات الأرض (FR-4):')
print(f'  🏜️  أرض مفتوحة : {n_open/total_pixels*100:.1f}%')
print(f'  🌳 مناطق خضراء : {n_green/total_pixels*100:.1f}%')
print(f'  🏢 مناطق مبنية : {n_built/total_pixels*100:.1f}%')

print('\n✅ التقسيم اكتمل!')
```

```python
# ═══════════════════════════════════════════════
# CELL 9 — ما بعد المعالجة (Post-processing)
# ═══════════════════════════════════════════════

def postprocess_mask(mask: np.ndarray,
                     opening_size: int = 3,
                     closing_size: int = 5,
                     min_area: int = 100) -> np.ndarray:
    """
    يُطبّق العمليات المورفولوجية لتنظيف القناع.

    الخطوات:
    1. Opening  → يُزيل الضوضاء الصغيرة (يُصغّر ثم يُكبّر)
    2. Closing  → يُغلق الثقوب الداخلية (يُكبّر ثم يُصغّر)
    3. فلترة المناطق الصغيرة جداً
    """
    mask_uint8 = mask.astype(np.uint8) * 255

    # 1. Opening: يُزيل الضوضاء (نقاط صغيرة مبعثرة)
    kernel_open  = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (opening_size, opening_size))
    opened = cv2.morphologyEx(mask_uint8, cv2.MORPH_OPEN, kernel_open)

    # 2. Closing: يُغلق الثقوب الصغيرة داخل المبنى
    kernel_close = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (closing_size, closing_size))
    closed = cv2.morphologyEx(opened, cv2.MORPH_CLOSE, kernel_close)

    # 3. فلترة المناطق الصغيرة جداً
    # نُعلّم المكوّنات المترابطة ونحذف الصغيرة
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(closed, connectivity=8)

    clean_mask = np.zeros_like(closed)
    for label_id in range(1, num_labels):  # نبدأ من 1 لتجاهل الخلفية
        area = stats[label_id, cv2.CC_STAT_AREA]
        if area >= min_area:
            clean_mask[labels == label_id] = 255

    return clean_mask > 0  # نُرجع Boolean


def separate_overlapping_buildings(combined_mask: np.ndarray) -> List[np.ndarray]:
    """
    يُفصل المباني المتداخلة باستخدام Watershed Transform.

    الخوارزمية:
    1. نُطبّق Distance Transform لإيجاد مراكز المباني
    2. نستخدم Watershed لتقسيم المناطق المتلاصقة
    """
    mask_uint8 = combined_mask.astype(np.uint8) * 255

    # Distance transform: يحسب لكل بكسل بُعده عن أقرب حافة
    dist_transform = cv2.distanceTransform(mask_uint8, cv2.DIST_L2, 5)

    # نُطبّع ونحدد النقاط البعيدة عن الحواف (مراكز المباني)
    dist_norm = cv2.normalize(dist_transform, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    _, sure_fg = cv2.threshold(dist_norm, 0.5 * dist_norm.max(), 255, 0)

    # نُطبّق Watershed
    sure_fg   = np.uint8(sure_fg)
    unknown   = cv2.subtract(mask_uint8, sure_fg)

    _, markers = cv2.connectedComponents(sure_fg)
    markers    = markers + 1
    markers[unknown == 255] = 0

    # Watershed يحتاج صورة ملوّنة
    if len(combined_mask.shape) == 2:
        color_img = cv2.cvtColor(mask_uint8, cv2.COLOR_GRAY2BGR)
    else:
        color_img = combined_mask

    cv2.watershed(color_img, markers)

    # نُحوّل كل منطقة منفصلة لقناع مستقل
    separated_masks = []
    for m_id in np.unique(markers):
        if m_id <= 1:  # تجاهل الخلفية والحدود
            continue
        region_mask = (markers == m_id)
        if region_mask.sum() > 50:  # فلترة المناطق الصغيرة
            separated_masks.append(region_mask)

    return separated_masks


# ─── تطبيق Post-processing على كل الأقنعة ───
print('🔧 تطبيق Post-processing على الأقنعة...')
print('='*50)

cleaned_masks = {}

for category in ['buildings', 'vegetation', 'open_land']:
    raw_masks = all_seg_results.get(f'{category}_masks', [])

    emoji = {'buildings': '🏢', 'vegetation': '🌳', 'open_land': '🏜️'}[category]
    print(f'\n{emoji} معالجة {category} ({len(raw_masks)} أقنعة)...')

    # معاملات مختلفة لكل فئة
    params = {
        'buildings' : {'opening_size': 3, 'closing_size': 7,  'min_area': 200},
        'vegetation': {'opening_size': 5, 'closing_size': 11, 'min_area': 500},
        'open_land' : {'opening_size': 7, 'closing_size': 15, 'min_area': 1000},
    }[category]

    processed = []
    for mask in raw_masks:
        clean = postprocess_mask(mask, **params)
        if clean.sum() > params['min_area']:  # تجاهل الأقنعة الفارغة بعد التنظيف
            processed.append(clean)

    cleaned_masks[category] = processed
    removed = len(raw_masks) - len(processed)
    print(f'    قبل: {len(raw_masks)} أقنعة → بعد: {len(processed)} أقنعة (أُزيل {removed})')


# ─── فصل المباني المتداخلة ───
print('\n🏢 فصل المباني المتداخلة (Watershed)...')
all_building_masks = cleaned_masks.get('buildings', [])

if all_building_masks:
    # ندمج كل أقنعة المباني في قناع واحد
    h, w = all_building_masks[0].shape
    combined_buildings = np.zeros((h, w), dtype=bool)
    for m in all_building_masks:
        combined_buildings |= m

    # نُفصّل باستخدام Watershed
    separated = separate_overlapping_buildings(combined_buildings)
    print(f'  قبل الفصل: {len(all_building_masks)} مجموعة → بعد الفصل: {len(separated)} مبنى')
    cleaned_masks['buildings_separated'] = separated
else:
    print('  لا توجد مباني لفصلها')
    cleaned_masks['buildings_separated'] = []

print('\n✅ Post-processing اكتمل!')
```

```python
# ═══════════════════════════════════════════════
# CELL 10 — إسناد معرّفات فريدة (Building IDs)
# ═══════════════════════════════════════════════

@dataclass
class DetectedBuilding:
    """
    يمثّل مبنى واحداً مكتشفاً مع كل بياناته.
    هذه البنية تُستخدم لاحقاً من قِبَل فريق FR-5 (الارتفاع) وFR-6 (التصنيف).
    """
    building_id  : str          # معرف UUID فريد
    bbox         : List[int]    # [x1, y1, x2, y2] بالبكسل
    area_pixels  : int          # مساحة القناع بالبكسل
    confidence   : float        # درجة ثقة الكشف
    mask         : np.ndarray   # القناع Boolean (H, W)
    centroid     : Tuple[int,int]  # مركز المبنى (x, y)
    image_source : str = ''     # مسار الصورة الأصلية


def register_buildings(building_masks: List[np.ndarray],
                       building_boxes: np.ndarray,
                       building_scores: np.ndarray,
                       image_path: str = '') -> List[DetectedBuilding]:
    """
    يُسجّل كل مبنى بمعرّف فريد ويحفظ بياناته.

    إذا كانت الأقنعة مفصولة بـ Watershed، نستخدم bbox محسوبة من القناع نفسه.
    """
    buildings = []

    for i, mask in enumerate(building_masks):
        # 1. حساب Bounding Box من القناع مباشرة (أدق)
        rows = np.any(mask, axis=1)
        cols = np.any(mask, axis=0)

        if not rows.any() or not cols.any():
            continue  # قناع فارغ — نتجاهله

        rmin, rmax = np.where(rows)[0][[0, -1]]
        cmin, cmax = np.where(cols)[0][[0, -1]]
        bbox = [int(cmin), int(rmin), int(cmax), int(rmax)]

        # 2. حساب المساحة
        area_pixels = int(mask.sum())

        # 3. حساب المركز
        centroid_y, centroid_x = ndimage.center_of_mass(mask)
        centroid = (int(centroid_x), int(centroid_y))

        # 4. درجة الثقة — من DINO إذا متاحة، وإلا 1.0
        if i < len(building_scores):
            confidence = float(building_scores[i])
        else:
            confidence = 1.0

        # 5. إنشاء المعرف الفريد
        building_id = f'BLD-{str(uuid.uuid4())[:8].upper()}'

        # 6. حفظ القناع على الديسك
        mask_path = DIRS['masks'] / f'{building_id}.png'
        Image.fromarray(mask.astype(np.uint8) * 255).save(mask_path)

        building = DetectedBuilding(
            building_id  = building_id,
            bbox         = bbox,
            area_pixels  = area_pixels,
            confidence   = confidence,
            mask         = mask,
            centroid     = centroid,
            image_source = image_path,
        )
        buildings.append(building)

    return buildings


# ─── تسجيل المباني ───
print('🆔 تسجيل المباني بمعرّفات فريدة...')

# نستخدم الأقنعة المفصولة إذا توفّرت، وإلا الأصلية
final_building_masks  = (
    cleaned_masks['buildings_separated']
    if cleaned_masks['buildings_separated']
    else cleaned_masks['buildings']
)

registered_buildings = register_buildings(
    building_masks  = final_building_masks,
    building_boxes  = all_seg_results.get('buildings_boxes', np.array([])),
    building_scores = detection_results['buildings']['scores'],
    image_path      = test_image_path,
)

# ─── حفظ البيانات كـ JSON ───
buildings_json = []
for b in registered_buildings:
    buildings_json.append({
        'building_id'  : b.building_id,
        'bbox'         : b.bbox,
        'area_pixels'  : b.area_pixels,
        'confidence'   : round(b.confidence, 4),
        'centroid'     : b.centroid,
        'mask_path'    : str(DIRS['masks'] / f'{b.building_id}.png'),
        'image_source' : b.image_source,
    })

with open(DIRS['results'] / 'buildings_registry.json', 'w', encoding='utf-8') as f:
    json.dump(buildings_json, f, indent=2, ensure_ascii=False)

# ─── عرض الملخص ───
print(f'\n📊 ملخص المباني المكتشفة:')
print(f'  عدد المباني   : {len(registered_buildings)}')
if registered_buildings:
    areas = [b.area_pixels for b in registered_buildings]
    print(f'  أكبر مبنى     : {max(areas):,} بكسل')
    print(f'  أصغر مبنى     : {min(areas):,} بكسل')
    print(f'  متوسط المساحة : {int(np.mean(areas)):,} بكسل')

    print(f'\n📋 أول 5 مباني:')
    print(f'  {"ID":<20} {"المساحة":>10} {"الثقة":>8} {"المركز":>15}')
    print('  ' + '-'*55)
    for b in registered_buildings[:5]:
        print(f'  {b.building_id:<20} {b.area_pixels:>10,} {b.confidence:>8.3f} {str(b.centroid):>15}')

print(f'\n✅ تم حفظ سجل المباني في: results/buildings_registry.json')
```

```python
# ═══════════════════════════════════════════════
# CELL 11 — التقييم والمقاييس
# ═══════════════════════════════════════════════

def compute_iou(pred_mask: np.ndarray, gt_mask: np.ndarray) -> float:
    """
    يحسب IoU بين قناع متوقع وقناع الحقيقة الأرضية.

    IoU = التقاطع / الاتحاد
    القيمة: 0 (لا تطابق) → 1 (تطابق تام)
    """
    pred = pred_mask.astype(bool)
    gt   = gt_mask.astype(bool)

    intersection = np.logical_and(pred, gt).sum()
    union        = np.logical_or(pred, gt).sum()

    if union == 0:
        return 1.0 if intersection == 0 else 0.0

    return float(intersection) / float(union)


def compute_pixel_metrics(pred_mask: np.ndarray, gt_mask: np.ndarray) -> dict:
    """
    يحسب مقاييس على مستوى البكسل:
    - Precision: من كل البكسلات التي توقعناها مبنى، كم منها صحيح؟
    - Recall: من كل البكسلات التي هي مبنى فعلاً، كم اكتشفنا؟
    - F1: الوسط الهارموني بين Precision و Recall
    """
    pred = pred_mask.astype(bool)
    gt   = gt_mask.astype(bool)

    TP = np.logical_and(pred,  gt).sum()   # كشف صحيح
    FP = np.logical_and(pred, ~gt).sum()   # كشف خاطئ (توقعنا مبنى ولا يوجد)
    FN = np.logical_and(~pred, gt).sum()   # فُقدان (مبنى موجود لم نكشفه)
    TN = np.logical_and(~pred, ~gt).sum()  # تجاهل صحيح

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    recall    = TP / (TP + FN) if (TP + FN) > 0 else 0.0
    f1        = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0
    iou       = compute_iou(pred_mask, gt_mask)

    return {
        'precision': round(precision, 4),
        'recall'   : round(recall, 4),
        'f1'       : round(f1, 4),
        'iou'      : round(iou, 4),
        'TP': int(TP), 'FP': int(FP), 'FN': int(FN)
    }


def evaluate_on_dataset(data_split: list, n_samples: int = 10) -> dict:
    """
    يُقيّم النظام على عدد من الصور.
    """
    all_iou, all_precision, all_recall, all_f1 = [], [], [], []
    detection_counts = []

    samples = random.sample(data_split, min(n_samples, len(data_split)))

    for sample in tqdm(samples, desc='جاري التقييم'):
        try:
            img_path  = sample['image']
            mask_path = sample['mask']

            gt_mask = np.array(Image.open(mask_path)) > 128

            # كشف
            det_res, img_src = run_dino_detection(img_path, BEST_PROMPTS, THRESHOLDS)
            boxes  = det_res['buildings']['boxes']
            scores = det_res['buildings']['scores']

            detection_counts.append(len(boxes))

            if len(boxes) == 0:
                all_iou.append(0.0)
                all_precision.append(0.0)
                all_recall.append(0.0)
                all_f1.append(0.0)
                continue

            # NMS
            if len(scores) > 0:
                keep = apply_nms(boxes, scores)
                boxes = boxes[keep]

            # تقسيم
            masks = run_sam_segmentation(img_src, boxes)

            if not masks:
                all_iou.append(0.0)
                all_precision.append(0.0)
                all_recall.append(0.0)
                all_f1.append(0.0)
                continue

            # دمج أقنعة المباني
            h, w = gt_mask.shape
            pred_combined = np.zeros((h, w), dtype=bool)
            for m in masks:
                if m.shape == (h, w):
                    pred_combined |= m

            # حساب المقاييس
            metrics = compute_pixel_metrics(pred_combined, gt_mask)
            all_iou.append(metrics['iou'])
            all_precision.append(metrics['precision'])
            all_recall.append(metrics['recall'])
            all_f1.append(metrics['f1'])

        except Exception as e:
            print(f'  ⚠️ خطأ في الصورة: {e}')
            continue

    return {
        'mean_iou'       : round(np.mean(all_iou), 4),
        'mean_precision' : round(np.mean(all_precision), 4),
        'mean_recall'    : round(np.mean(all_recall), 4),
        'mean_f1'        : round(np.mean(all_f1), 4),
        'avg_detections' : round(np.mean(detection_counts), 2),
        'n_evaluated'    : len(all_iou),
    }


# ─── تشغيل التقييم ───
print('📏 جاري التقييم على Test Set...')
print('='*50)

eval_results = evaluate_on_dataset(test_data, n_samples=8)

print('\n📊 نتائج التقييم:')
print(f'  الصور المقيَّمة   : {eval_results["n_evaluated"]}')
print(f'  Mean IoU          : {eval_results["mean_iou"]:.4f}')
print(f'  Mean Precision    : {eval_results["mean_precision"]:.4f}')
print(f'  Mean Recall       : {eval_results["mean_recall"]:.4f}')
print(f'  Mean F1 Score     : {eval_results["mean_f1"]:.4f}')
print(f'  متوسط الكشوفات   : {eval_results["avg_detections"]:.1f} مبنى/صورة')

# ─── تفسير النتائج ───
print('\n📌 تفسير المقاييس:')
iou = eval_results['mean_iou']
if iou >= 0.7:
    print('  ✅ أداء ممتاز (IoU ≥ 0.7)')
elif iou >= 0.5:
    print('  🟡 أداء جيد (0.5 ≤ IoU < 0.7)')
elif iou >= 0.3:
    print('  🟠 أداء مقبول (0.3 ≤ IoU < 0.5)')
else:
    print('  🔴 أداء ضعيف — جرّب تحسين الـ Prompts أو العتبات')
    print('  💡 تذكر: النتائج أفضل مع صور جوية حقيقية عالية الدقة')

# حفظ نتائج التقييم
with open(DIRS['results'] / 'evaluation_results.json', 'w') as f:
    json.dump(eval_results, f, indent=2)

print('\n✅ التقييم اكتمل!')
```

```python
# ═══════════════════════════════════════════════
# CELL 13 — وضع الاستنتاج (Inference Mode)
# ═══════════════════════════════════════════════

from google.colab import files
import io

def run_full_pipeline(image_input, save_results=True):
    """
    يُشغّل خط المعالجة الكامل على صورة واحدة ويعرض النتائج.
    """
    print('Starting full pipeline...')
    import time
    start_time = time.time()

    # ── تحميل الصورة ──
    if isinstance(image_input, str):
        image_np = np.array(Image.open(image_input).convert('RGB'))
        img_path = image_input
    else:
        image_np = np.array(image_input.convert('RGB'))
        img_path = '/tmp/user_uploaded.jpg'
        Image.fromarray(image_np).save(img_path)

    # تقليص الصورة إذا كانت كبيرة (لتسريع المعالجة)
    max_dim = 1024
    oh, ow = image_np.shape[:2]
    if max(oh, ow) > max_dim:
        ratio = max_dim / max(oh, ow)
        image_np = np.array(
            Image.fromarray(image_np).resize(
                (int(ow * ratio), int(oh * ratio)), Image.LANCZOS
            )
        )
        Image.fromarray(image_np).save(img_path)

    h, w = image_np.shape[:2]
    print(f'  Image size: {w} x {h} px')

    # ── خطوة 1: كشف DINO ──
    print('  [1/7] Detection (Grounding DINO)...')
    det_results, img_src = run_dino_detection(img_path, BEST_PROMPTS, THRESHOLDS)
    total_det = sum(len(det_results[c]['boxes']) for c in det_results)
    print(f'        {total_det} raw detections')

    # ── خطوة 2: NMS ──
    print('  [2/7] NMS filtering...')
    seg_input = {}
    for cat in ['buildings', 'vegetation', 'open_land']:
        boxes  = det_results[cat]['boxes']
        scores = det_results[cat]['scores']
        if len(boxes) > 0 and len(scores) > 0:
            keep  = apply_nms(boxes, scores, iou_threshold=0.45)
            boxes = boxes[keep]
        seg_input[f'{cat}_boxes'] = boxes

    # ── خطوة 3: SAM ──
    print('  [3/7] Segmentation (SAM)...')
    for cat in ['buildings', 'vegetation', 'open_land']:
        masks = run_sam_segmentation(img_src, seg_input[f'{cat}_boxes'])
        seg_input[f'{cat}_masks'] = masks
        print(f'        {cat}: {len(masks)} masks')

    # ── خطوة 4: Post-processing ──
    print('  [4/7] Post-processing...')
    params_map = {
        'buildings' : {'opening_size': 3, 'closing_size': 7,  'min_area': 150},
        'vegetation': {'opening_size': 5, 'closing_size': 11, 'min_area': 400},
        'open_land' : {'opening_size': 7, 'closing_size': 15, 'min_area': 800},
    }
    clean_seg = {}
    for cat in ['buildings', 'vegetation', 'open_land']:
        raw   = seg_input[f'{cat}_masks']
        clean = [postprocess_mask(m, **params_map[cat]) for m in raw]
        clean = [m for m in clean if m.sum() > params_map[cat]['min_area']]
        clean_seg[f'{cat}_masks'] = clean
        clean_seg[f'{cat}_boxes'] = seg_input[f'{cat}_boxes']

    # ── خطوة 5: خريطة الأرض ──
    print('  [5/7] Land use map...')
    global land_use_map
    land_use_map = create_land_use_map(image_np.shape, clean_seg)

    # ── خطوة 6: فصل المباني ──
    print('  [6/7] Separating buildings...')
    bld_masks = clean_seg['buildings_masks']
    if bld_masks:
        comb = np.zeros((h, w), dtype=bool)
        for m in bld_masks:
            if m.shape == (h, w):
                comb |= m
        final_bld = separate_overlapping_buildings(comb)
        if not final_bld:
            final_bld = bld_masks
    else:
        final_bld = []

    # ── خطوة 7: Building IDs ──
    print('  [7/7] Assigning building IDs...')
    buildings = register_buildings(
        final_bld,
        seg_input['buildings_boxes'],
        det_results['buildings']['scores'],
        img_path
    )

    elapsed = time.time() - start_time

    # ── ملخص ──
    total_px  = land_use_map.size
    print(f'\n  Results ({elapsed:.1f}s):')
    print(f'    Buildings detected : {len(buildings)}')
    print(f'    Open land          : {np.sum(land_use_map==0)/total_px*100:.1f}%')
    print(f'    Vegetation         : {np.sum(land_use_map==1)/total_px*100:.1f}%')
    print(f'    Built-up           : {np.sum(land_use_map==2)/total_px*100:.1f}%')

    # ── التصور المحسَّن ──
    save_p = str(DIRS['vis'] / 'inference_result.png') if save_results else None

    visualize_results(image_np, clean_seg, buildings, det_results, save_p)

    return {
        'image'       : image_np,
        'buildings'   : buildings,
        'land_use_map': land_use_map,
        'seg_results' : clean_seg,
    }


# ─── واجهة الرفع ───
print('Upload your aerial image:')
print('Supported: JPG, PNG, TIFF')
print()

uploaded = files.upload()

if uploaded:
    filename   = list(uploaded.keys())[0]
    img_bytes  = uploaded[filename]
    user_image = Image.open(io.BytesIO(img_bytes)).convert('RGB')
    print(f'Uploaded: {filename}  ({user_image.width}x{user_image.height})')
    results = run_full_pipeline(user_image)
    print('\nDone!')
else:
    print('No file uploaded — running on test image...')
    results = run_full_pipeline(test_data[0]['image'])
```
