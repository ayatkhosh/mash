```python
# ============================================================================
# الخلية 5️⃣: فئات تحميل الصور والمعالجة
# ============================================================================

class DataLoader:
    """تحميل ومعالجة الصور"""

    def __init__(self, target_size=(512, 512)):
        self.target_size = target_size
        self.images = []
        self.image_names = []

    def load_images(self, folder_path):
        """تحميل صور من مجلد"""
        supported = ('.png', '.jpg', '.jpeg', '.tiff', '.tif', '.bmp')

        for filename in os.listdir(folder_path):
            if filename.lower().endswith(supported):
                img_path = os.path.join(folder_path, filename)

                img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
                if img is None:
                    print(f"⚠️ لم يتمكن من تحميل: {filename}")
                    continue

                # تحويل BGR إلى RGB
                if len(img.shape) == 3 and img.shape[2] == 3:
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # تغيير الحجم
                img = cv2.resize(img, self.target_size)

                self.images.append(img)
                self.image_names.append(filename)
                print(f"✅ تم تحميل: {filename}")

        return self.images, self.image_names

# اختبار DataLoader
loader = DataLoader()
images, image_names = loader.load_images(INPUT_DIR)
print(f"\n✅ تم تحميل {len(images)} صورة بنجاح!")
```

```python
# ============================================================================
# الخلية 8️⃣: معالجة الصور (Post-Processing)
# ============================================================================

class PostProcessor:
    """معالجة الأقنعة والصور"""

    @staticmethod
    def remove_small_objects(mask, min_area=100):
        """إزالة الأشياء الصغيرة"""
        mask = (mask > 0).astype(np.uint8)
        num_labels, labels_im = cv2.connectedComponents(mask)

        result = np.zeros_like(mask)
        for i in range(1, num_labels):
            component_mask = (labels_im == i).astype(np.uint8)
            area = np.sum(component_mask)
            if area >= min_area:
                result = np.logical_or(result, component_mask).astype(np.uint8)

        return result

    @staticmethod
    def fill_holes(mask):
        """ملء الفراغات الصغيرة"""
        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
        mask = cv2.morphologyEx((mask > 0).astype(np.uint8), cv2.MORPH_CLOSE, kernel)
        return mask

    @staticmethod
    def apply_morphological_ops(mask):
        """عمليات مورفولوجية"""
        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
        mask = cv2.morphologyEx((mask > 0).astype(np.uint8), cv2.MORPH_OPEN, kernel)
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
        return mask

processor = PostProcessor()
print("✅ تم إنشاء معالج ما بعد المعالجة!")
```

```python
# ============================================================================
# الخلية 9️⃣: توليد معرفات المباني
# ============================================================================

class BuildingIDGenerator:
    """توليد معرفات فريدة للمباني"""

    def __init__(self, prefix="BLD"):
        self.prefix = prefix
        self.buildings = []
        self.counter = 0

    def generate_ids(self, masks, image_name=None):
        """توليد معرفات من الأقنعة"""
        buildings_info = []

        for mask in masks:
            self.counter += 1
            building_id = f"{self.prefix}_{self.counter:06d}"

            # حساب المساحة
            area = np.sum(mask > 0)

            # حساب المركز
            y_coords, x_coords = np.where(mask > 0)
            if len(x_coords) > 0:
                center_x = float(np.mean(x_coords))
                center_y = float(np.mean(y_coords))
            else:
                center_x = center_y = 0

            building_info = {
                'building_id': building_id,
                'image_name': image_name,
                'area': float(area),
                'center_x': center_x,
                'center_y': center_y,
                'timestamp': datetime.now().isoformat()
            }

            self.buildings.append(building_info)
            buildings_info.append(building_info)

        return buildings_info

    def save_to_json(self, output_path):
        """حفظ النتائج في JSON"""
        with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(self.buildings, f, indent=4, ensure_ascii=False)
        print(f"✅ تم الحفظ في: {output_path}")

# إنشاء مثيل من البناء
building_generator = BuildingIDGenerator()
print("✅ تم إنشاء مولد المعرفات!")
```
